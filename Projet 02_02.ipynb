```python
data6_pays = data6.value_counts('country').to_frame().sort_values(by=['country'])
data6_pays.reset_index(level=0, inplace=True)
data6_pays = data6_pays.rename({0:'number_of_universities'}, axis=1)
data6_pays = data6_pays.replace({'United States of America': 'United States'})
data6_pays.loc[:,'Ratio'] = 'value'
data6_pays
```

```python
data3_higher = data3[data3['institute_type'].str.contains('Higher Education Institutions')]
data3_higher_all = data3_higher[data3_higher['direct_expenditure_type'].str.contains('Total')]
data3_higher_all[['country', 'institute_type', '2011']]

data_merge = data3_higher_all.merge(data6_pays, how='inner', on='country')
df = data_merge[['country','2011','number_of_universities','Ratio']]
df
```

```python
df.loc[:,'Ratio']=df['2011']/df['number_of_universities']
df.sort_values('Ratio').dropna().reset_index(drop=True)[['country','Ratio']]
```
